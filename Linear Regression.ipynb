```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some random data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3.5, 5.3, 7.1, 9.2])
```

```python
# Calculate the slope and y-intercept using linear algebra
X = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(X, y, rcond=None)[0]
```

```python
# Plot the data points and the regression line
plt.scatter(x, y)
plt.plot(x, m*x + c, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Print the slope and y-intercept
print('Slope:', m)
print('Y-Intercept:', c)
```
